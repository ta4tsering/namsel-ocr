```python
import re
import yaml

from bs4 import BeautifulSoup
from collections import defaultdict
from pathlib import Path
from pyewts import pyewts
from strsimpy.cosine import Cosine
from uuid import uuid4

converter = pyewts()
```

```python
def to_yaml(dict_):
    return yaml.safe_dump(dict_, sort_keys = False, allow_unicode=True)
```

```python
def from_yaml(yml_path):
    return yaml.safe_load(yml_path.read_text(encoding='utf-8'))
```

```python
def reconstruct_text(text_id, text, title):
    print(text_id)
    cur_text[text_id] = {
        'pedurma_title': text['title'],
        'text_title': title[1],
        'rkts_id': title[0],
        'img_loc_start': text['img_loc_start'],
        'img_loc_end': text['img_loc_end'],
        'pg_start': text['pg_start'],
        'pg_end': text['pg_end'],
        'vol': text['vol'],
        'durchen': text['durchen'],    
    }
    return cur_text
```

```python
old_outline = from_yaml(Path('./kangyur_text_durchen.yml'))
```

```python
pedurma_text_titles = from_yaml(Path('./pedurma_text_title.yml'))
```

```python
len(old_outline)
```

```python
new_outline = {}
for (text_id, text), title in zip(old_outline.items(), pedurma_text_titles):
    cur_text = {}
    cur_text = reconstruct_text(text_id, text, title)
    new_outline.update(cur_text)
```

```python
new_outline_yml = to_yaml(new_outline)
```

```python
Path('./new_pedurma_outline.yml').write_text(new_outline_yml, encoding='utf-8')
```
